```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skewnorm
```

```python
# range harga jual
x = np.linspace(3000, 10000, 500)
```

```python
# rescale parameters
a = 6
loc = 3.5 * 1000     # shift loc
scale = 1.5 * 1000   # widen scale
```

```python
# probability density, right tail skew
y = skewnorm.pdf(x, a=a, loc=loc, scale=scale) * 1000

np.random.seed(10) # masukkan NPM
y = y + np.random.normal(0, 0.005, size=x.shape)  # add Gaussian noise
y = y*1000
```

```python
# harga produksi
produksi_y= np.linspace(0, 1000, len(x))
produksi_z = np.linspace(3000, 2000, len(x)) + np.random.normal(0, 0.1, size=x.shape)*100
# linearly decreases from 3000 -> 2000 for quantity 0 -> 1000
```

```python
# Plot
plt.figure(figsize=(8,5))
plt.plot(x, y, color="navy", label='penjualan', linewidth=2)
plt.plot(produksi_z, produksi_y, color="red", label='produksi', linewidth=2)
plt.title("Toko Roti Maknyus")
plt.xlabel("x (harga)")
plt.ylabel("y (jumlah barang dibuat/laku)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Hitung Profit
q = np.minimum(y, produksi_y)    # jumlah laku
revenue = x * q                  # pendapatan
cost = produksi_z * produksi_y   # biaya produksi total
profit = revenue - cost          # keuntungan
```

```python
# Cari Keuntungan Maksimum
idx_max = np.argmax(profit)
q_optimal = q[idx_max]
harga_optimal = x[idx_max]
profit_max = profit[idx_max]
```

```python
print("------ JAWABAN ------")
print(f"Jumlah produksi optimal: {q_optimal:.0f} roti")
print(f"Harga jual optimal: {harga_optimal:.2f}")
print(f"Keuntungan maksimum: {profit_max:.2f}")
```

```python
# plot profit
plt.figure(figsize=(8,5))
plt.plot(produksi_y, profit, color="blue", linewidth=2, label="Profit")
plt.axvline(x=produksi_y[idx_max], color="red", linestyle="--", linewidth=2, label="Produksi Optimal")
plt.scatter(produksi_y[idx_max], profit_max, color="black", zorder=5)
plt.title("Keuntungan Toko Roti")
plt.xlabel("Jumlah Produksi")
plt.ylabel("Profit")
plt.legend()
plt.grid(True)
plt.show()
```
